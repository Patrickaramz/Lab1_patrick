```python
import matplotlib.pyplot as plt

#ppnar filen och lser in alla rader
with open("dna_raw.txt", "r") as file:
    lines = file.readlines()

# skapar en tom dictionary fr att spara sekvenserna
sequences = {}
current_id = "" 


# Gr igenom varje rad i filen
for line in lines:
    line = line.strip()  #Tar bort extra mellanrum

    # OM raden r ett ID (allts har >)
    if line.startswith(">"):
        current_id = line[1:]  # tar bort > symbol
        sequences[current_id] = ""  
    else:
        # Lgger till sjlva sekvensen (gr den till sm bokstver fr att inte blanda A och a)
        sequences[current_id] += line.lower()

# Gr igenom varje sekvens och rknar hur mnga av varje bokstav som finns
for id, sequence in sequences.items():
    letter_count = {}

    # Rknar varje bokstav i sekvensen
    for letter in sequence:
        if letter in letter_count:
            letter_count[letter] += 1
        else:
            letter_count[letter] = 1

    # Skriver ut ID, sekvensen och hur mnga av varje bokstav som finns
    print(f"\n{id}")
    print(sequence)
    print(letter_count)

    sequence_number = id.replace("seq", "")
    display_name = f"DNA sequence {sequence_number}"

    # Hr ritar jag ett stapeldiagram med resultatet
    plt.bar(letter_count.keys(), letter_count.values(), color="blue")
    plt.title(display_name) # Titeln fr varje sekvens
    plt.xlabel("DNA-Letters") # x-axeln visar bokstverna (a, g, t, osv)
    plt.ylabel("Amount") # y-axeln visar antalet
    plt.tight_layout() # Ser till att allt fr plats och att inget krockar i grafen
    plt.show()
```
